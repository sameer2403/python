List : ordered & mutable

```python
#creating a list of fruits
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits)
```

```python
for fruit in fruits:
    print(fruit)
```

```python
for i in range(len(fruits)):
    print(f"Index: {i}, Fruit: {fruits[i]}")
```

```python
n= len(fruits)
print(f"Number of fruits: {n}")
for i in range(n):
    print(f"Index: {i}, Fruit: {fruits[i]}")
```

```python
a = [1, 2, 3, 4, 5, True]
```

```python
a.append(6)
print(a)
```

```python
a.remove(3)
```

Tuple : ordered & immutable

```python
colors = ("red", "green", "blue", "yellow", "purple")
print(colors)
print(type(colors))
```

```python
colors[0] = "orange"
```

```python
for i in colors:
    print(i)
```

set : unordered & changeable, 
does not allow duplicate value

```python
students = {"Alice": 85, "Bob": 90, "Charlie": 78}
print(students)
print(type(students))
```

```python
students["David"] = 92
print(students)
```

```python
students.pop("Bob")
print(students)
```

```python
set1 = {1, 2, 3,4, 4, 5}
print(set1)
```

dictionary: key:value pair
ordered -> by key
indexable -> by key

```python
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
```

```python
person["age"]
```

```python
person["age"] = 31
print(person)
```

```python
person["city"] = "Los Angeles"
print(person)
```

```python
person["is_student"] = False
print(person)
```

```python
#loop dictionary
for key in person:
    print(f"{key}: {person[key]}")
```

```python
for key, value in person.items():
    print(f"{key}: {value}")
```
