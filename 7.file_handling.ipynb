**1.Opening & reading file**

```python
#open a file in read mode
file = open("notes.txt", "r")

#read the contents of the file
content = file.read()
print(content)

#close the file
file.close()
```

2. using with to read(auto- closing)

```python
with open("notes.txt", "r") as file:
    content = file.read()
    print(content)
```

**3.writing a ile**

```python
with open("outputs.txt", "w") as file:
    file.write("This is a new line.\n")
```

```python
with open("outputs.txt", "w") as file:
    file.write("This is another line.\n")
```

```python
with open("outputs.txt", "a") as file:
    file.write("This is second line.\n")
```

```python
with open("outputs.txt", "r") as file:
    for line in file:
        print(line)
```

```python
with open("outputs.txt", "r") as file:
    for line in file:
        print(line.strip())
```

2. function with parameters

```python
def greet(name):
    print(f"Hello, {name}!")

```

```python
greet("Alice")
```

```python
def sum(a, b):
    return a + b

result = sum(5, 3)
print(result)
```

3. return a value

```python
def square(x):
    return x * x

# save result to a variable
result = square(5)
print(result)

print(square(10))
```

4. default parameters

```python
def add(a=100, b=200):
    return a + b
print(add())
print(add(10, 20))
```
